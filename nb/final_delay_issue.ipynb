# the issue

Executing programs using acquire_decimated with multiple reps seems to behave differently if final_delay is set to None versus if it is set to 0, even though they should be the same (?). When set to 0, the program triggers correctly on the first rep but is slightly off on reps after the first rep (if the program starts at t=0). When set to None, the program exhibits very strange behavior, shown below.

# test program that we will use to demonstrate this problem

```python
class const_pulse(AveragerProgramV2):
    def _initialize(self, cfg):
        ro_ch = cfg['ro_ch']
        gen_ch = cfg['gen_ch']
        
        self.declare_gen(ch=gen_ch, nqz=2, mixer_freq=cfg['freq'])
        self.declare_readout(ch=ro_ch, length=cfg['ro_len'])
        self.add_readoutconfig(ch=ro_ch, name='ro', freq=cfg['freq'], gen_ch=gen_ch)

        self.add_pulse(ch=gen_ch, name="pulse1", ro_ch=ro_ch, 
                       style="const", 
                       length=cfg['pulse_len'],
                       freq=cfg['freq'], 
                       phase=0,
                       gain=cfg['gain'], 
                      )
        self.add_pulse(ch=gen_ch, name="pulse2", ro_ch=ro_ch, 
                       style="const", 
                       length=cfg['pulse_len'],
                       freq=cfg['freq'], 
                       phase=0,
                       gain=cfg['gain']/2, 
                      )

    def _body(self, cfg):
        self.send_readoutconfig(ch=cfg['ro_ch'], name='ro', t=0)
        self.pulse(ch=cfg['gen_ch'], name="pulse1", t=0)
        self.pulse(ch=cfg['gen_ch'], name="pulse2", t=cfg['pulse_len'])
        self.trigger(ros=[cfg['ro_ch']], pins=[0], t=cfg['trig_time'], ddr4=True)
        

config = {'gen_ch': 8,
          'ro_ch': 1,
          'freq': 5000,
          'trig_time': 0.45,
          'ro_len': 3.1,
          'gain': 0.15,
          'pulse_len': 0.3
         }
```

# final_delay=0
We see that, on reps after the first rep, the program triggers close to the correct time, but it is off by a little bit.

```python
prog = const_pulse(soccfg, reps=4, final_delay=0, cfg=config)
iq_list = prog.acquire_decimated(soc, soft_avgs=1)
t = prog.get_time_axis(ro_index=0)

fig, ax = plt.subplots(2, 2, figsize=(9, 7))
ax[0][0].plot(t, iq_list[0][0])
ax[0][0].set_title("rep 1")
ax[0][1].plot(t, iq_list[0][1])
ax[0][1].set_title("rep 2")
ax[1][0].plot(t, iq_list[0][2])
ax[1][0].set_title("rep 3")
ax[1][1].plot(t, iq_list[0][3])
ax[1][1].set_title("rep 4")
```

However, if we modify our program so that the first pulse is sent only after some time has passed, instead of at t=0, we see that everything is as we would expect. This is a bit strange.

```python
class const_pulse_delayed(AveragerProgramV2):
    def _initialize(self, cfg):
        ro_ch = cfg['ro_ch']
        gen_ch = cfg['gen_ch']
        
        self.declare_gen(ch=gen_ch, nqz=2, mixer_freq=cfg['freq'])
        self.declare_readout(ch=ro_ch, length=cfg['ro_len'])
        self.add_readoutconfig(ch=ro_ch, name='ro', freq=cfg['freq'], gen_ch=gen_ch)

        self.add_pulse(ch=gen_ch, name="pulse1", ro_ch=ro_ch, 
                       style="const", 
                       length=cfg['pulse_len'],
                       freq=cfg['freq'], 
                       phase=0,
                       gain=cfg['gain'], 
                      )
        self.add_pulse(ch=gen_ch, name="pulse2", ro_ch=ro_ch, 
                       style="const", 
                       length=cfg['pulse_len'],
                       freq=cfg['freq'], 
                       phase=0,
                       gain=cfg['gain']/2, 
                      )

    def _body(self, cfg):
        self.send_readoutconfig(ch=cfg['ro_ch'], name='ro', t=0)
        self.pulse(ch=cfg['gen_ch'], name="pulse1", t=1) # not 0
        self.pulse(ch=cfg['gen_ch'], name="pulse2", t=1+cfg['pulse_len'])
        self.trigger(ros=[cfg['ro_ch']], pins=[0], t=cfg['trig_time'], ddr4=True)
        

config_delayed = {'gen_ch': 8,
          'ro_ch': 1,
          'freq': 5000,
          'trig_time': 1.45, # previously 0.45
          'ro_len': 3.1,
          'gain': 0.15,
          'pulse_len': 0.3
         }

prog = const_pulse_delayed(soccfg, reps=4, final_delay=0, cfg=config_delayed)
iq_list = prog.acquire_decimated(soc, soft_avgs=1)
t = prog.get_time_axis(ro_index=0)

fig, ax = plt.subplots(2, 2, figsize=(9, 7))
ax[0][0].plot(t, iq_list[0][0])
ax[0][0].set_title("rep 1")
ax[0][1].plot(t, iq_list[0][1])
ax[0][1].set_title("rep 2")
ax[1][0].plot(t, iq_list[0][2])
ax[1][0].set_title("rep 3")
ax[1][1].plot(t, iq_list[0][3])
ax[1][1].set_title("rep 4")
```
